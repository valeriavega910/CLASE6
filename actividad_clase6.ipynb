1. Calculadora de IMC (Índice de Masa Corporal):

Cree una función que calcule el IMC (Índice de Masa Corporal) de una persona a partir de su peso y altura. La función debe recibir como parámetros el peso (en kg) y la altura (en cm) y devolver el valor del IMC.

Opcional: Implemente una segunda función que clasifique el IMC según los estándares de la OMS:

Bajo peso: IMC < 18.5
Peso normal: 18.5 <= IMC < 25
Sobrepeso: 25 <= IMC < 30
Obesidad: IMC >= 30

```python
def IMC(peso,altura_ingresada):
    altura = altura_ingresada / 100
    imc = peso / (altura ** 2)
    imc_redondeado = round(imc,2)
    if imc < 18.5:
        print("El valor del IMC es de ",imc_redondeado,"= tiene bajo peso")
    elif 18.5 <= imc < 25:
        print("El valor del IMC es de ",imc_redondeado,"= tiene peso normal")
    elif 25 <= imc < 30:
        print("El valor del IMC es de ",imc_redondeado,"= tiene sobrepeso")
    else:
        imc >= 30
        print("El valor del IMC es de ",imc_redondeado,"= tiene obesidad")
            

        
```

```python
IMC(92,172)
```

2. Analizador de números:

Cree una función que reciba un número entero como parámetro y devuelva un diccionario con la siguiente información:

Si el número es par o impar
Si el número es positivo, negativo o cero
Si el número es divisible por 3
Si el número es divisible por 5
Si el número es divisible por 3 y por 5
Opcional: Implemente la función utilizando una función lambda para cada una de las verificaciones.

```python
def diccionario_analizador(numero):

    par=lambda valor_entero:valor_entero % 2 == 0
    impar=lambda valor_entero:valor_entero % 2 != 0
    positivo=lambda valor_entero:valor_entero > 0
    negativo=lambda valor_entero:valor_entero < 0 
    divisible_3=lambda valor_entero:valor_entero % 3 == 0
    divisible_5=lambda valor_entero:valor_entero % 5 == 0
    divisible_3y5=lambda valor_entero:valor_entero % 3 == 0 and valor_entero % 5 == 0

    diccionario = {
        "par":par(numero),
        "impar":impar(numero),
        "positivo":positivo(numero),
        "negativo":negativo(numero),
        "divisible_3":divisible_3(numero),
        "divisible_5":divisible_5(numero),
        "divisible_3y5":divisible_3y5(numero)
        }
    return diccionario
```

```python
diccionario_analizador(19)
```

3. Calculadora de área y perímetro de figuras geométricas:

Cree funciones para calcular el área y perímetro de las siguientes figuras geométricas:
Cuadrado
Triángulo (equilátero, rectángulo e isósceles)
Círculo
Rectángulo

Cada función debe recibir las dimensiones necesarias como parámetros y devolver el valor correspondiente del área o perímetro.

```python
def cuadrado(lado1):
    area=lado1**2
    perimetro=lado1*4
    return area,perimetro
```

```python
cuadrado(5)
```

```python
def triang_equilatero(lado,altura):
    area=(lado*altura)/2
    perimetro=lado*3
    return area,perimetro
```

```python
triang_equilatero(7,5)
```

```python
def triang_rectangulo(base,altura,hipotenusa):
    area=(base*altura)/2
    area_redondeado=round(area,2)
    perimetro=base+altura+hipotenusa
    return area_redondeado,perimetro
```

```python
triang_rectangulo(5.2,3,6)
```

```python
def triang_isosceles(base,altura_base,lado):
    area=(base*altura_base)/2
    area_redondeado=round(area,2)
    perimetro=(lado*2)+base
    return area_redondeado,perimetro
```

```python
triang_isosceles(8,6,5)
```

```python
def circulo(radio):
    area=3.1416*radio**2
    area_redondeado=round(area,2)
    perimetro=2*3.1416*radio
    perimetro_redondeado=round(perimetro,2)
    return area_redondeado,perimetro_redondeado
```

```python
circulo(2)
```

```python
def rectangulo(lado1,lado2):
    area=lado1*lado2
    perimetro=2*(lado1+lado2)
    return area,perimetro
```

```python
rectangulo(7,4)
```
